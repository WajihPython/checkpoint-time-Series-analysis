```python
import matplotlib.pyplot as plt
def plot_timeseries(df, x, y, title="", xlabel='Month', ylabel='Passengers', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_timeseries(series, x=series.index, y=series['#Passengers'], title='Passengers per month')
```
